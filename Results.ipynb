```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Study 1

```python
eff1 = pd.read_csv('Study1_effnet.csv')
res1 = pd.read_csv('Study1_resnet.csv')
vgg1 = pd.read_csv('Study1_vgg.csv')
```

```python
study1 = pd.concat([eff1,res1,vgg1],axis=0)
```

```python
study1 =study1.reset_index(drop=True)
study1
```

Precision Plot

```python
N = 5
ind = np.arange(N)
width = 0.25

x_vals = list(res1['Precision'].values)
bar1 = plt.bar(ind, x_vals, width, color = 'r')

y_vals = list(eff1['Precision'].values)
bar2 = plt.bar(ind+width, y_vals, width, color = 'b')

z_vals = list(vgg1['Precision'].values)
bar3 = plt.bar(ind+width*2, z_vals, width, color = 'g')

plt.xlabel("Classes")
plt.ylabel('Precision')
plt.title("Precision Values")

plt.xticks(ind+width,['Mild', 'Moderate', 'No DR', 'PDR', 'Severe'])
plt.legend( (bar1, bar2, bar3), ('ResNet', 'EfficientNet', 'VGG16') )
plt.show()
```

Recall Plot

```python
N = 5
ind = np.arange(N)
width = 0.25

x_vals = list(res1['Recall'].values)
bar1 = plt.bar(ind, x_vals, width, color = 'r')

y_vals = list(eff1['Recall'].values)
bar2 = plt.bar(ind+width, y_vals, width, color = 'b')

z_vals = list(vgg1['Recall'].values)
bar3 = plt.bar(ind+width*2, z_vals, width, color = 'g')

plt.xlabel("Classes")
plt.ylabel('Precision')
plt.title("Precision Values")

plt.xticks(ind+width,['Mild', 'Moderate', 'No DR', 'PDR', 'Severe'])
plt.legend( (bar1, bar2, bar3), ('ResNet', 'EfficientNet', 'VGG16') )
plt.show()
```
